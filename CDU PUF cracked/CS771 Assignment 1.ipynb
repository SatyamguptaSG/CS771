```python
import numpy as np
x = np.genfromtxt('train_challenges.dat')
y = np.genfromtxt('train_responses.dat')
indices = np.random.permutation(len(x))
split_ratio = 0.70
split_index = int(len(x) * split_ratio)
x_train, x_test , y_train , y_test = x[indices[:split_index]], x[indices[split_index:]],y[indices[:split_index]],y[indices[split_index:]]
```

```python
d=2048
w=np.zeros((2,512))
start=sum(y)/(512*2048)
w[0,:512]=np.repeat(start,512)
for i in range(512): 
    w[1][i]=i
```

```python
def itr(w,x,split_index,y,n):
    temp=np.zeros(512)
    for i in range(split_index):
        temp+=((w@x[i,:512])-y[i])*x[i]
    w=w-n*temp
    #print(w)
    return w
```

```python
def error(w,x,y):
    n=len(y_test)
    err=0
    for i in range(0,n,1):
        err+=pow((w@x[i]-y[i]),2)
    err=pow(err,0.5)
    err/=(2*n)
```

```python
n=0.001 #n=step length
iterations=10000
for i in range(iterations):
    print(w[0])
    w[0]=itr(w[0],x_train,split_index,y_train,n)
    w=w[ :,w[0].argsort()]
    #for i in range(512,2048,1):
     #   w[0][i]=0
    #w=w[ :,w[1].argsort()]
    print(w[0])
```
